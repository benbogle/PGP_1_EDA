```python
def explore_datasets(df_dict):
    """
    Function to print the shape, info, description and number of missing values of a particular group of datasets.
    
    Args:
        df_dict (dict): Dictionary containing the name and pandas dataframe object of the datasets to explore
    
    Returns: None    
    """
    
    for dataset, df in df_dict.items():
        
        print(f'Exploring {dataset} dataset...\n')
        
        print(f'Shape of {dataset} dataset = {df.shape}\n')
        
        print(f'Info about {dataset} dataset:\n')
        df.info()
        print('')
        
        print(f'Description of {dataset} dataset:')
        display(df.describe())
        print('')
        
        print(f'Number of missing values in {dataset} dataset = {df.isnull().sum().sum()}')
        print('----------------------------------------------------------------------------\n')
        
```

### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use `sum(axis=1)` store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.  

```python
# Convert genre columns to type int
for col in movies.columns[3:]:
    movies[col] = movies[col].astype(int)
```

```python
# Create aggregation dictionary to pass to groupby function so we know how to aggregate each column
# Count 'movie title' column to get total number of releases per year, and sum all genre columns to get number of releases in each year per genre
agg_dict = {
    col: 'count' if col == 'movie title' else 'sum' for col in movies.drop(columns=['movie id', 'release date', 'unknown']).columns
}

# Aggregate movies dataset by year and apply aggregations
yearly = (
movies    
    .groupby(movies['release date'].dt.year)
    .agg(agg_dict)
    .rename(columns={'movie title': 'total'}) # rename aggregated column to 'total' - represents total movies released that year
)

# View final few rows of dataframe to check output
yearly.tail(3)
```

```python
# Get percentage view to calculate the percent of number of releases in a year for each genre
# (I believe Hint 2 doesn't make sense, as using sum(axis=1) would give you the total number of rated genres in a year, not the total number of movie releases, since a movie can have more than one genre)

yearly_pct = np.round(yearly.div(yearly['total'], axis=0) * 100, 1)

# Check last 10 rows of data
yearly_pct.tail(10)
```

```python
# Plot heatmap. Have transposed the dataframe as I think it is easier to read rowwise
plt.figure(figsize=(20,20))
sns.heatmap(yearly_pct.drop(columns=['total']).T, cmap="rocket_r")
plt.show()
```

**Insights:**

1. Action films were very rare until the mid-sixties, accounting for just 3% of films per year on average before 1966. From 1966 onwards, their popularity increased and they accounted for 18% of films per year on average (code calculations not shown). This is also illustrated by the darker colours to the right of the row on the above heatmap
2. Musicals appear to have become less popular. In the early-fifties we can see they accounted for a large portion of films released at the time, denoted by the darker colours. However, from 1985 onwards they have accounted for a far smaller share of total releases
3. Drama films have always been universally popular and have remained relevant over the whole period, accounting for 37% of films per year on average (code calculation not shown)

### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
# Create a new dataframe from the ratings dataset at the movie id level and aggregate to pull total ratings and average rating for each movie
# Merge to movies dataset to get movie title and rearrange columns
# Filter out movies with 100 reviews or fewer
# Sort by avg ratings descending and select top 25

top_movies_25 = (
ratings
    .groupby('movie id')
    .agg( # aggregate
        total_ratings=('rating', 'count'),
        avg_rating=('rating', 'mean'))
    .merge( # merge to get movie title
        movies[['movie id', 'movie title']],
        left_index=True,
        right_on='movie id')
    [['movie title', 'movie id', 'total_ratings', 'avg_rating']] # rearrange columns
    .query('total_ratings > 100') # filter out movies with <= 100 ratings
    .sort_values('avg_rating', ascending=False) # sort by avg_rating, highest to lowest
    .head(25) # select top 25
    .round(2) # round rating for easier readability
    .reset_index(drop=True) # reset index to start from 0
)

top_movies_25
```

### 10. Check for the validity of the below statements with respect to the data provided - 10 marks

* Men watch more drama than women

* Men watch more Romance than women

**compare the percentages**

1. Merge all the datasets

2. There is no need to conduct statistical tests around this. Just **compare the percentages** and comment on the validity of the above statements.

3. you might want ot use the .sum(), .div() function here.

4. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

```python
# Merge all datasets together and inspect object
# Each row represents a rating given by a particular user on a specific film (which may have more than one genre)
data = (
pd
    .merge(users, ratings, on='user id')
    .merge(movies, on='movie id')
)
data.head(3)
```

```python
# Get list of all genres
genres = movies.columns[4:].values

# Create agg_dict to pass to groupby function
agg_dict = {col: 'sum' for col in genres}
agg_dict['rating'] = 'count'

# Groupby gender, aggregate and rearrange columns
gender = (
data
    .groupby('gender')
    .agg(agg_dict)
    .rename(columns={'rating': 'total_ratings'})
    [['total_ratings'] + list(genres)]
)

print('This dataframe shows the number of ratings given by each gender and genre pair')
gender.head()
```

```python
# Create a new dataframe showing ratings by each genre as a % of total ratings, split by gender
gender_pct = gender.div(gender.total_ratings, axis=0).round(2)
gender_pct.head()
```

```python
# Check validity of statements
gender_pct[['Drama', 'Sci-Fi', 'Romance']]
```
